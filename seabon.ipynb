```python
data=sb.load_dataset("iris")
```

```python
data
```

```python
# print sb.get_dataset_names()
```

```python
# allows ustoplot pairwise relationship between variables
sb.pairplot(data)
```

```python
# shows relationship between data from a specified column
sb.pairplot(data,hue="species")
```

```python
# forcing marginal histograms
sb.pairplot(data,hue="species",diag_kind="hist")
```

```python
# displays KDE kernel density estimation
sb.pairplot(data,kind="kde")
```

```python
g=sb.relplot(kind="scatter",data=data,x="petals",y="petal")
g.set(title='Petal Sepal relationship')
```

```python
g = sb.relplot(kind='scatter', data=data, x='sepal_width', y='petal_width')
g.set(title='Sepal Width Petal width relationship')
```

```python
g = sb.relplot(kind='scatter', data=data, x='sepal_width', y='petal_width',hue="species")
g.set(title='Sepal_width Petal_width relationship')
```
